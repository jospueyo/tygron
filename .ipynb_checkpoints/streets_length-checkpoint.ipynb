# Get the total length of streets in a bbox

```python
from sklearn.neighbors import KDTree
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import networkx as nx
import numpy as np
import osmnx as ox
import pandas as pd
import geopandas
import pygeoj

%matplotlib inline
```

```python
#Import bbox coordinates
bbox = pygeoj.load("./Girona_02/bbox_gi.geojson")

north = bbox.bbox[3]
south = bbox.bbox[1]
east = bbox.bbox[2]
west = bbox.bbox[0]

print(north, south, east, west)
```

```python
#Create a graph by city name (Girona)
G = ox.graph_from_bbox(north, south, east, west, network_type = 'drive')
# quick plot
ox.plot_graph(G, bgcolor='white', figsize=(10,10), edge_color='black')
```

```python
#undirect graph to avoid double edges
G = ox.utils_graph.get_undirected(G)
```

```python
edges = ox.graph_to_gdfs(G, nodes=False)
print(sum(edges['length']))
```
